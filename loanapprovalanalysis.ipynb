# Loan Approval Analysis

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('loan_approval_dataset.csv')
df.head(5)
df.columns = df.columns.str.strip()
df['loan_status'] = df['loan_status'].str.strip()
df.head(5)
```

```python
print(df.query("loan_status == 'Approved'")["income_annum"].mean())
print(df.query("loan_status == 'Rejected'")["income_annum"].mean())

```

Income analysis

```python

df.groupby('loan_status')['income_annum'].mean().plot(kind='bar')
plt.xlabel('Loan Status')
plt.ylabel('Mean Annual Income')
plt.title('Average Annual Income by Loan Status')
plt.show()
print(df.query("loan_status == 'Approved'")["income_annum"].mean())
print(df.query("loan_status == 'Rejected'")["income_annum"].mean())
print("Salary has little to no effect on loan approval.")
```

Cibil Score Impact

```python
df.head(5)
df.groupby('loan_status')['cibil_score'].mean().plot(kind='bar')
plt.title('Average CIBIL Score by Loan Status')
plt.xlabel('Loan Status')
plt.ylabel('Mean CIBIL Score')
plt.show()
print("Higher CIBIL scores are associated with loan approval.")
```

Approval Rate by education

```python
sns.countplot(x='education', hue='loan_status', data=df)
plt.title('Education Level Distribution by Loan Status')
print("Education level does not appear to have a significant impact on loan approval.")
```
